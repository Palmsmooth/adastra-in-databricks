## Reading Data
The extraction part is easy--ones can read almost any file formats in PySpark!


### CSV

```python
from pyspark import SparkFiles
```

```python
WORLD_BANK_POPULATION_CSV_URL = 'https://raw.githubusercontent.com/AdastraTH/2024-univ-workshop/main/datasets/populations/population.csv'

spark.sparkContext.addFile(WORLD_BANK_POPULATION_CSV_URL)
populations_data = spark.read.csv("file://" + SparkFiles.get("population.csv"), inferSchema=True, header=True)
display(populations_data)
```


### JSON

```python
COUNTRY_CATEGORIES_JSON_URL = 'https://raw.githubusercontent.com/AdastraTH/2024-univ-workshop/main/datasets/populations/country_categories.json'

spark.sparkContext.addFile(COUNTRY_CATEGORIES_JSON_URL)
country_categories_data = spark.read.json("file://" + SparkFiles.get("country_categories.json"), multiLine=True)
display(country_categories_data)
```


## Storing data

Now as we have read the data into DataFrames (fancy name for tables), we should write it to the "data store" for future permanent retrieval

```python
%sql

CREATE DATABASE IF NOT EXISTS silver
```

```python
populations_data.write.mode("overwrite").saveAsTable("silver.populations")
```
