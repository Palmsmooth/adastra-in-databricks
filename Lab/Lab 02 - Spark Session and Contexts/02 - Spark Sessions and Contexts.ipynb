
In Databricks, when a Notebook is executed, it will creates a `spark` object which is a **Spark Session**.

```python
%py

spark
```


It is this Spark session that is the heart of Spark during any executions. Loosely speaking, a notebook can be viewed as a **Spark Application**, sharing a single Spark Session and therefore sharing any data read by it.

Let's try reading some sample data. **You don't have to understand anything except this is reading the data and storing into something called `iris_data`**

```python
%py

sc.addFile("https://raw.githubusercontent.com/domoritz/maps/master/data/iris.json")
iris_data = spark.read.option("multiLine", "true").json(f"file://{SparkFiles.get('iris.json')}")
iris_data.createOrReplaceTempView("iris_data")
```

```python
%py

display(iris_data)
```


We've wrote Python above, but since the context is shared, we can actually use other languages to continue what we have done, like we can switch to SQL immediately!

```python
%sql

SELECT * FROM iris_data WHERE species = 'setosa'
```
