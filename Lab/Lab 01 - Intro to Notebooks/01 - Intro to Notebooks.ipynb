## What are Notebooks?

Notebooks are a notebook-like development environment where you can write codes in parts and run it in small chunks.

This helps in some development aspects (for example, data science), where you may want to do some tasks and spend time checking intermediary results.


## Components of a Notebook

The only single component you have to care about is **Cells**. Cells are blocks like this, where you will run code as a whole in a single cell.

Databricks Notebooks' Cells should start a line with a **magic comment** indicating the language of the contents in the cell. For example

- `%md` stands for Markdown, for writing texts and not scripts.
- `%py` stands for Python.
- `%sql` stands for SQL.

If there are no magic comments, the notebook falls back to its default language, in this case Python.

Let's try writing some Python code, in a cell beginning with Python magic.

```python
%py

def hello(name):
  print(f"Hello, {name}!")
```

Cells are continuous and the context of the cells are shared in between a single notebook. Let's try calling the function we created earlier in another cell. This time without Python magic (to show that by default Python code is expected!)

```python
hello("University Workshop")
```
